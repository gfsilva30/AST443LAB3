```python
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.nddata import CCDData
from astropy.visualization import hist
from ccdproc import *
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
```

```python
# read in dark files
dark0 = CCDData.read('lab3_dark2min.00000026.DARK.FIT', unit='adu')
dark1 = CCDData.read('lab3_dark2min.00000027.DARK.FIT', unit='adu')
dark2 = CCDData.read('lab3_dark2min.00000028.DARK.FIT', unit='adu')
dark3 = CCDData.read('lab3_dark2min.00000029.DARK.FIT', unit='adu')
dark4 = CCDData.read('lab3_dark2min.00000030.DARK.FIT', unit='adu')
dark5 = CCDData.read('lab3_dark2min.00000031.DARK.FIT', unit='adu')
dark6 = CCDData.read('lab3_dark2min.00000032.DARK.FIT', unit='adu')

# combining darks
combined_darks = Combiner([dark0, dark1, dark2, dark3, dark4, dark5, dark6])
master_dark = combined_darks.median_combine()

#Making new FIT file from median array
hdu = fits.PrimaryHDU(master_dark)
hdul = fits.HDUList([hdu])
hdu.writeto('master_dark2min.DARK.FIT', overwrite=True)

#prints dark image
dark_image2min = CCDData.read('master_dark2min.DARK.FIT', unit='adu')
plt.imshow(dark_image2min.data, cmap = 'viridis', vmin= np.mean(master_dark.data) - 3*np.std(master_dark.data), vmax = np.mean(master_dark.data) + 3*np.std(master_dark.data))
plt.colorbar()
```

```python
#Plot Master Dark Image Histogram
dark0 = dark_image2min.data.flatten()
hist(dark0, bins=100, label='Dark Image 2min')
plt.grid()
plt.legend()
plt.xlabel('Pixel value')
plt.ylabel('Number of pixels')
plt.semilogy() #y-axis in log-scale
```

```python
# read in flat field files
# flat0 = CCDData.read('lab3_flats.00000003.FIT', unit='adu')
flat1 = CCDData.read('lab3_flats_120sec.00000001.FIT', unit='adu')
flat2 = CCDData.read('lab3_flats_120sec.00000003.FIT', unit='adu')
flat3 = CCDData.read('lab3_flats_120sec.00000004.FIT', unit='adu')
flat4 = CCDData.read('lab3_flats_120sec.00000005.FIT', unit='adu')
flat5 = CCDData.read('lab3_flats_120sec.00000006.FIT', unit='adu')

# combining flats
combiner_flats = Combiner([flat1, flat2, flat3, flat4, flat5])
combined_median_flats = np.array(combiner_flats.median_combine())

# creating a master normed flat
master_flat0 = (combined_median_flats - dark_image30sec.data)
mean_flats = np.mean(master_flat0)
master_flat = master_flat0 / mean_flats

# subtracting the dark from the flat
# final_flat = master_flat - dark_image2min

hdu_flats = fits.PrimaryHDU(master_flat)
hdul_flats = fits.HDUList([hdu_flats])
hdu_flats.writeto('mas_flat.FIT', overwrite=True)

#plots image
flat_image = CCDData.read('mas_flat.FIT', unit='adu')
plt.imshow(flat_image.data, cmap='viridis', vmin = np.mean(master_flat) - 3*(np.std(master_flat)), vmax= np.mean(master_flat) + 3*(np.std(master_flat)))
plt.colorbar()
```

```python
#Cutting out the images to keep only the data from the target slit of 50µm
#ds9 was used to determine the physical cut
flat_spectro = flat_image.data[112:185, :]
np.shape(flat_spectro)

#Making new FIT file from median array
hdu_flats2 = fits.PrimaryHDU(flat_spectro)
hdul_flats2 = fits.HDUList([hdu_flats2])
hdu_flats2.writeto('cut_spec.FIT', overwrite=True)

#Displaying the cut spectrum image
spec_image = CCDData.read('cut_spec.FIT', unit='adu')

#Finding the Standard Deviation and Mean for vmin and vmax in order to show the image
std_spec = np.std(spec_image.data)
mean_spec = np.mean(spec_image.data)

#Displaying the cut spectrum image
plt.imshow(spec_image.data, cmap='viridis', vmin=mean_spec-3*std_spec, vmax=mean_spec+3*std_spec)
plt.colorbar()
```

```python
#Flatten the spectrum data and plot the flat-field values against the pixel positions
spec_graph = np.median(spec_image.data, axis=0)
x_pos = np.linspace(1,765,765)
plt.plot(x_pos, spec_graph)
plt.xlabel('Pixel Positions along the X-Axis')
plt.ylabel('Flat-Field Values')
```

```python
#Line of Best Fit Calibration
a, b, c, d = np.polyfit(x_pos, spec_graph, deg=3)
xseq = np.linspace(1, 765, 765)
plt.plot(x_pos, spec_graph)
plt.plot(x_pos, a * xseq**3 + b*xseq**2 + c*xseq**1 + d, color="black", lw=1.5, label=f'{np.around(a,15)}*x^3 + {np.around(b,10)}*x^2 + {np.around(c,10)}*x + {np.around(d,5)}')
plt.xlabel('Pixel Positions along the X-Axis')
plt.ylabel('Flat-Field Values')
plt.legend()
```
