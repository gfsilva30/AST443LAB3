```python
# reading in the cut normed images of gamma geminorum
gamma_gem0 = CCDData.read('cut_gamgem_normed0.FIT', unit='adu')
gamma_gem1 = CCDData.read('cut_gamgem_normed1.FIT', unit='adu')
gamma_gem2 = CCDData.read('cut_gamgem_normed2.FIT', unit='adu')
gamma_gem3 = CCDData.read('cut_gamgem_normed3.FIT', unit='adu')
gamma_gem4 = CCDData.read('cut_gamgem_normed4.FIT', unit='adu')

# combining gamma gem by flattening and then summing
flatten_gg0 = np.mean(gamma_gem0.data, axis=0)
flatten_gg1 = np.mean(gamma_gem1.data, axis=0)
flatten_gg2 = np.mean(gamma_gem2.data, axis=0)
flatten_gg3 = np.mean(gamma_gem3.data, axis=0)
flatten_gg4 = np.mean(gamma_gem4.data, axis=0)

gamma_gem_matrix = np.stack((flatten_gg0, flatten_gg1, flatten_gg2, flatten_gg3, flatten_gg4))
mean_gg = np.mean(gamma_gem_matrix, axis=0)
```

```python
# calibrating the wavelengths
calibrated_wavelengths = np.array([])
xseq3 = np.linspace(0, 765, 765)
for i in range(len(xseq3)):
    calibrated_wavelengths = np.append(calibrated_wavelengths, a + b*xseq3[i])

# plotting the wavelength versus intensity counts
plt.plot(calibrated_wavelengths, mean_gg)
a1, b1, c1 = np.polyfit(calibrated_wavelengths, mean_gg, deg=2)
plt.plot(calibrated_wavelengths, a1*calibrated_wavelengths**2 + b1*calibrated_wavelengths + c1, color="k", lw=1.5, label=f"{np.round(a1,4)}*x^2 + {np.round(b1,4)}*x+ {np.round(c1,5)}")
plt.vlines(5889.95, np.min(mean_gg), np.max(mean_gg), color='red', label='Sodium Absorption Line')
plt.vlines(6558, np.min(mean_gg), np.max(mean_gg), color='cyan', label=r'H$\alpha$ Absorption Line')

plt.xlabel('Wavelength ($\AA$)')
plt.ylabel('Intensity Counts')
plt.legend()
```

Perfect Black-Body Distribution:
$$B_{\nu}(\lambda, T) = \frac{2 h c^2}{\lambda^5} \frac{1}{e^{hc/(\lambda k_{B} T)} - 1}$$

```python
# creating a function for a perfect black body distribution
h = 6.62607015e-34
c = 3e8
kB = 1.380649e-23
T = 9260.0
def black_body(l):
    return ((2*h*(c**2))/(l**5)) * (1/(np.exp( (h*c)/((l/1e10)*kB*T))-1))
```

```python
# creating a ratio of the blackbody to the calibrated wavelengths
ratio = black_body(calibrated_wavelengths) / (a1*calibrated_wavelengths**2 + b1*calibrated_wavelengths + c1)
```

```python
# reading in the cut normed images of Vesta
v0 = CCDData.read('cut_vesta_normed0.FIT', unit='adu').data[17:68, :]
v1 = CCDData.read('cut_vesta_normed1.FIT', unit='adu').data[38:66, :]
v2 = CCDData.read('cut_vesta_normed2.FIT', unit='adu').data[29:64,:]
v3 = CCDData.read('cut_vesta_normed3.FIT', unit='adu').data[50:70, :]
v4 = CCDData.read('cut_vesta_normed4.FIT', unit='adu').data[58:72, :]
v5 = CCDData.read('cut_vesta_normed5.FIT', unit='adu').data[48:71, :]
v6 = CCDData.read('cut_vesta_normed6.FIT', unit='adu').data[41:59, :]

# mean combining the vesta images
flatten_v0 = np.mean(v0, axis=0)
flatten_v1 = np.mean(v1, axis=0)
flatten_v2 = np.mean(v2, axis=0)
flatten_v3 = np.mean(v3, axis=0)
flatten_v4 = np.mean(v4, axis=0)
flatten_v5 = np.mean(v5, axis=0)
flatten_v6 = np.mean(v6, axis=0)

vesta_matrix = np.stack((flatten_v0, flatten_v1, flatten_v2, flatten_v3, flatten_v4, flatten_v5, flatten_v6))
mean_v = np.mean(vesta_matrix, axis=0)
```

```python
# plotting the wavelength versus intensity counts for Vesta
plt.plot(calibrated_wavelengths, mean_v*ratio)
plt.xlabel('Wavelength ($\AA$)')
plt.ylabel('Intensity Counts')

#comparing absorption of vesta to sun's absorption lines (Fraunhofer lines)
plt.vlines(5889.95, np.min(mean_v*ratio), np.max(mean_v*ratio), color='red', label='Sodium Absorption Line')
plt.vlines(6558, np.min(mean_v*ratio), np.max(mean_v*ratio), color='cyan', label=r'H$\alpha$ Absorption Line')
plt.legend()
```
